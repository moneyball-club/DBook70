# Finding Player and Game Id

```python
from nba_api.stats.static import players
import pandas
pandas.set_option('display.max_colwidth',250)
pandas.set_option('display.max_columns', None)
```

```python
nba_players = players.get_players()
d_book = [player for player in nba_players
                   if player['full_name'] == 'Devin Booker'][0]
d_book
```

```python
from nba_api.stats.static import teams
nba_teams = teams.get_teams()
suns = [team for team in nba_players
                   if team['abbreviation'] == 'PHX'][0]
suns
```

```python
suns_id = suns['id']
```

```python
from nba_api.stats.endpoints import leaguegamefinder
gamefinder = leaguegamefinder.LeagueGameFinder(team_id_nullable=suns_id)
games = gamefinder.get_data_frames()[0]
```

```python
games_1617 = games[games.SEASON_ID.str[-4:] == '2016']
bos_games_1617 = games_1617[games_1617.MATCHUP.str.contains('BOS')]
bos_games_1617.head()
```
